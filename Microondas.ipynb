# Definção do Automata!

```python
from automathon import DFA

Q = {'Desligado (Sem Timer)', 'Desligado (Com Timer)', 'Porta Aberta (Sem Timer)', 'Porta Aberta (Com Timer)', 'Ligado'}
sigma = {'Muda Potência', 'Botão Pré-Programado', 'Adiciona Timer', 'Abre Porta', 'Liga', 'Remove Timer', 'Fecha Porta', 'Desliga', 'Fim Timer'}
delta = {
    'Desligado (Sem Timer)': {
        'Muda Potência': 'Desligado (Sem Timer)',
        'Botão Pré-Programado': 'Desligado (Com Timer)',
        'Adiciona Timer': 'Desligado (Com Timer)',
        'Abre Porta': 'Porta Aberta (Sem Timer)'
    },
    'Desligado (Com Timer)': {
        'Muda Potência': 'Desligado (Com Timer)',
        'Botão Pré-Programado': 'Desligado (Com Timer)',
        'Abre Porta': 'Ligado',
        'Liga': 'Ligado',
        'Remove Timer': 'Desligado (Sem Timer)'
    },
    'Porta Aberta (Sem Timer)':{
        'Fecha Porta': 'Desligado (Sem Timer)'
    },
    'Porta Aberta (Com Timer)':{
        'Fecha Porta': 'Desligado (Sem Timer)',
    },
    'Ligado':{
        'Abre Porta': 'Porta Aberta (Com Timer)',
        'Desliga': 'Desligado (Com Timer)',
        'Fim Timer': 'Desligado (Sem Timer)'
    }
}
initial_state = 'Desligado (Sem Timer)'
F = {'Desligado (Sem Timer)', 'Desligado (Com Timer)'}

automata = DFA(Q, sigma, delta, initial_state, F)
```

```python
automata.is_valid()
```

# Exemplo de Aplicação
Digite quantos comandos quiser ao terminar basta digitar um "enter" vazio

```python
command = []

while True:
    command.append(input())
    if not command[-1]:
        break

automata.accept(command[:-1])
```
